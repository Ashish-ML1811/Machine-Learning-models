Task B:

K-Nearest Neighbor Classification

For performing KNN classification we need to import the required libraries like 'KNeighborClassifier' and 'cross_val_score' for cross validation. We have performed the KNN classification using the values of K ranging from 1- 51 with the 10 fold cross validation which can be executed as follows;

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

k_values = list(range(1, 51))
KNN_f1_scores = []
KNN_acc_scores = []

for k in k_values:
    KNN = KNeighborsClassifier(n_neighbors = k, n_jobs=-1)
    f1_scores = cross_val_score(KNN, HAR_training_dataset, HAR_training_labels, cv=10, scoring="f1_weighted", n_jobs=-1)
    acc_scores = cross_val_score(KNN, HAR_training_dataset, HAR_training_labels, cv=10, scoring="accuracy", n_jobs=-1)
    
    KNN_f1_scores.append(f1_scores.mean())
    KNN_acc_scores.append(acc_scores.mean())
    
    print("The F1 scores of KNN on Train dataset:{}".format(np.round(KNN_f1_scores, 8)))
    print("The Accuracy score of KNN on Train dataset:{}".format(np.round(KNN_acc_scores, 8)))
```

Once the F1 and Accuracy values of the KNN model are determined we would like to determine the best values for Kwhich can be performed as follows; 

```python
best_k = k_values[KNN_f1_scores.index(max(KNN_f1_scores))]
print("best k value:", best_k)
```

Plot for the Cross validation with repect to K can be performed as follows

```python
import matplotlib.pyplot as plt
import matplotlib

plt.plot(range(1, 51), KNN_acc_scores)
plt.ylabel('Accuracy')
plt.xlabel('K- values')
```

Once the best value of K has been determined we will evaluate the performance of the model by substituting the value of K with its best value as 10, determining the testing and training accuracy, which can be performed as follows;

```python
from sklearn import metrics

# Build a KNN using 5 neighbor nodes
KNN_model = KNeighborsClassifier(n_neighbors = 10)

KNN_model.fit(HAR_training_dataset, HAR_training_labels)

KNN_accuracy = metrics.accuracy_score(HAR_training_labels, KNN_model.predict(HAR_training_dataset))
print("KNN training accuracy:{}".format(KNN_accuracy))

KNN_accuracy_test = metrics.accuracy_score(HAR_testing_labels, KNN_model.predict(HAR_testing_dataset))
print("KNN testing accuracy:{}".format(KNN_accuracy_test))
```

Reporting the confusion matrix for the KNN model which shows thevalues filled diagonally and also some values in various other rows and columns.

```python
from sklearn.metrics import confusion_matrix

KNN_confusion = metrics.confusion_matrix(HAR_training_labels, KNN_model.predict(HAR_training_dataset))
print('Confusion Matrix')
print(KNN_confusion)
```

```python
from sklearn.metrics import classification_report

KNN_classification_report = metrics.classification_report(HAR_testing_labels, KNN_model.predict(HAR_testing_dataset))
print("Classification reports")
print(KNN_classification_report)
```

Task C:

Multiclass Logistic Regression with Elastic Net:

For determining the Multiclass Logistic Regression using Elastic Net we need to first develop the classifier using SGD classifier, further we performed the 10 fold cross validation using various combinations of alpha and l1_ratio values, and choosing the best values out of it.

```python
# Multiclass logistic regression with Elastic net

from sklearn.linear_model import SGDClassifier
from sklearn.grid_search import GridSearchCV

alpha = [1e-4, 3e-4, 1e-4, 3e-3, 1e-2, 3e-2]
l1 = [0, 0.15, 0.5, 0.7, 1]
pars = dict(alpha=alpha, l1_ratio = l1)
enet = SGDClassifier(loss='log', penalty='elasticnet')
Score = GridSearchCV(enet, pars, cv=10)


enet.fit(HAR_training_dataset, HAR_training_labels)
```

```python
Score.fit(HAR_training_dataset, HAR_training_labels)
```

One the classifier has been developed and once the training dataset has been fitted over it we can further calculate the means of the accurcay and f1 scores for the given training dataset. Upon using command as GridSearchCV, result(Accuracy and f1 values) has been generated by default and stored in the object 'result'. Hence, upon using the command 'mean_validation_score' over the default result we get the means of the accuracy and f1 scores as follows;

```python
Score_means = [result.mean_validation_score for result in Score.grid_scores_]
print(Score_means)
```

Once the means of the Accuracy and f1 scores has been generated we will calculate the best Alpha and l1_ratio values has been calculated as follows;

```python
print("The best values for Alpha and l1_ratios will be like:{}".format(Score.best_params_))
```

Once the best values for alpha and l1_ratio values are generated we will further plot the surface plot for the same as follows;

```python
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D


# surface plot of F1 score

in_n =0
a = alpha # the possible options for hyperparameter (variable) 1
b = l1

mean= np.zeros([6,5])
for i in range(0,6):
    for j in range(0,5):
        mean[i,j]=Score_means[in_n]
        in_n=in_n+1
        
X,Y  = np.meshgrid(b, a)

C = mean
print(X)
print(Y)
print(mean)
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, C, rstride=1, cstride=1, antialiased=True)
ax.set_xlabel('L1-Ratio')
ax.set_ylabel('Alpha Values')
ax.set_zlabel('F1-Score')
plt.show()
```

Once we get the best values for the alpha and l1_ratio we will retrain the model using the respective best values to get the model accuracy over training and testing datasets as follows;

```python
from sklearn import metrics
# Build a SGD using 5 neighbor nodes
enet = SGDClassifier(penalty='elasticnet', alpha=0.003, l1_ratio=0.5)

# Fit the model using our training data
enet.fit(HAR_training_dataset, HAR_training_labels)

# Training Accuracy
enet_accuracy = metrics.accuracy_score(HAR_training_labels, enet.predict(HAR_training_dataset))
print("Model training accuracy:{}".format(enet_accuracy))

enet_accuracy_test = metrics.accuracy_score(HAR_testing_labels, enet.predict(HAR_testing_dataset))
print("Model testing accuracy:{}".format(enet_accuracy_test))
```

Once the model is retrained using the best values we will calculate the confusion matrix for the above model, which again shows some values diagonally also some other values in other rows and columns.

```python
from sklearn.metrics import confusion_matrix

enet_confusion = metrics.confusion_matrix(HAR_training_labels, enet.predict(HAR_training_dataset))
print('Confusion Matrix')
print(enet_confusion)
```

```python
from sklearn.metrics import classification_report

enet_report = metrics.classification_report(HAR_testing_labels, enet.predict(HAR_testing_dataset))
print("Classification reports")
print(enet_report)
```

Task D:

Support Vector Machine (RBF Kernel)

For determining the Support Vector Machine (RBF Kernel) we need to first develop the classifier using SGD SVC classifier, further we performed the 10 fold cross validation using various combinations of gamma and C_values, and choosing the best values out of it.

```python
from sklearn import svm
from sklearn.grid_search import GridSearchCV

gamma_values = [1e-3, 1e-4]
C_values = [1, 10, 100, 1000]
SVM_par = dict(gamma=gamma_values, C=C_values)
SVM_model = svm.SVC(kernel='rbf')
SVM_grid = GridSearchCV(SVM_model, SVM_par, cv=10, scoring="f1_weighted")

SVM_model.fit(HAR_training_dataset, HAR_training_labels)
SVM_grid.fit(HAR_training_dataset,  HAR_training_labels)
```

One the classifier has been developed and once the training dataset has been fitted over it we can further calculate the means of the accurcay and f1 scores for the given training dataset. Upon using command as GridSearchCV, result(Accuracy and f1 values) has been generated by default and stored in the object 'result'. Hence, upon using the command 'mean_validation_score' over the default result we get the means of the accuracy and f1 scores as follows;

```python
SVM_model_means = [result.mean_validation_score for result in SVM_grid.grid_scores_]
print(SVM_model_means)
```

Once the best values for gamma and C values are generated we will further plot the surface plot for the same as follows;

```python
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D

in_n =0
g = gamma_values # the possible options for hyperparameter (variable) 1
c = C_values

mean= np.zeros([4,2])
for i in range(0,4):
    for j in range(0,2):
        mean[i,j]=SVM_model_means[in_n]
        in_n=in_n+1
        
X,Y  = np.meshgrid(g, c)

Z = mean
print(X)
print(Y)
print(Z)
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, antialiased=True)
ax.set_xlabel('Gamma Values')
ax.set_ylabel('C Values')
ax.set_zlabel('F1-Score')
plt.show()
```

Once we get the best values for the gamma and C values we will retrain the model using the respective best values to get the model accuracy over training and testing datasets as follows;

```python
print("The best values for C and gamma are as follows: {}".format(SVM_grid.best_params_))
```

```python
SVM_model = svm.SVC(kernel='rbf', C=1000, gamma=0.001)
SVM_model.fit(HAR_training_dataset, HAR_training_labels)

predict = SVM_model.predict(HAR_testing_dataset)

```

```python
SVM_accuracy = metrics.accuracy_score(HAR_testing_labels, predict)
print("The Model accuracy for SVM model will be:{}".format(SVM_accuracy))
```

Once the model is retrained using the best values we will calculate the confusion matrix for the above model, which again shows some values diagonally also some other values in other rows and columns.

```python
from sklearn.metrics import confusion_matrix

SVM_model_confusion = metrics.confusion_matrix(HAR_training_labels, SVM_model.predict(HAR_training_dataset))
print('Confusion Matrix')
print(SVM_model_confusion)
```

```python
from sklearn.metrics import classification_report

SVM_model_report = metrics.classification_report(HAR_testing_labels, SVM_model.predict(HAR_testing_dataset))
print("Classification reports")
print(SVM_model_report)
```

Random Forest:
